```python
# plotting function :
input_file = rcms[0]
import netCDF4 as nc 
import cartopy.crs as ccrs
import cartopy
def plot_errors_maps(data,lon, lat, minn,maxx,nn, cmap, name):

    import matplotlib as mpl
    fig = plt.figure('1')
    fig.set_size_inches(14, 10)

    ax = plt.axes(projection=ccrs.PlateCarree())
    v =  np.linspace(minn,maxx , nn, endpoint=True)
    #cmap = plt.cm.  # define the colormap
    cmap = cmap
    # extract all colors from the .jet map
    cmaplist = [cmap(i) for i in range(cmap.N)]
#    # force the first color entry to be grey
#    cmaplist[0] = (.5, .5, .5, 1.0)
    ax.coastlines('50m', linewidth=0.8)
    ax.add_feature(cartopy.feature.BORDERS)
    ax.add_feature(cartopy.feature.OCEAN,
                   edgecolor='black', zorder=0,
                   linewidth=0.8, alpha=.7)
    ax.add_feature(cartopy.feature.BORDERS,
                   edgecolor='black', zorder=0,
                   linewidth=0.8, alpha=.7)
    ax.add_feature(cartopy.feature.LAND, zorder=0,
                   linewidth=0.8, alpha=.7)
    # create the new map
    cmap = mpl.colors.LinearSegmentedColormap.from_list('Custom cmap', cmaplist, cmap.N)
    norm = mpl.colors.BoundaryNorm(v, cmap.N)
    cs = plt.pcolor(lon, lat, data, cmap=cmap,norm=norm,
                    linewidth=0,rasterized=True,shading="auto")
    ax.gridlines( draw_labels=True,linewidth=0, color='black', alpha=0.5, linestyle='--')
    #cs = plt.contourf(lons, lats, t, v, transform=ccrs.PlateCarree(), cmap=plt.cm.BuGn)
    cax = fig.add_axes([ax.get_position().x1+0.05,ax.get_position().y0,0.02,ax.get_position().height])
    cb = plt.colorbar(cs, cax=cax, extend="both")
    cb.set_label("mm/day", fontsize=20, rotation=270, labelpad=20)
    #ax.gridlines( draw_labels=True,dms=True,linewidth=2, color='black', alpha=0.5, linestyle='--')
    #cb.set_label('total precip [mm/day]', fontsize=20)
    cb.ax.tick_params(labelsize=20)

    from shapely.geometry import box
    # Define the bounding box values (lonmax, latmin, lonmin, latmax)

    #bbox = (92, 37 ,55, 50)
    bbox = (54, 35 ,50, 38)
    
    # Create a rectangular polygon from the bbox
    rect = box(*bbox)

    # Add the polygon to the map
    ax.add_geometries([rect], crs=ccrs.PlateCarree(), facecolor='none', edgecolor='k', linewidth=2)
    plt.savefig(name, format='png', bbox_inches='tight')
```

```python
# importing
data_dir ="/home/bijan/Documents/scripts/CCN_constrained_downscaling/DATA/"
rcms=['TOT_PREC_EC-Earth3-Veg_CCLM_2019_2033.nc']
gcms=['TOT_PREC_EC-Earth3-Veg_GCM_2019_2033.nc']
```

```python
import numpy as np 
def read_netcdf_files(input_file,output_file ,var,alpha=1, reduce=False):
    # this function reads the netcdf files and
    # converts them to pytorch data as in the prototype code
    import netCDF4 as nc 
    import numpy as np
    import torch
    import netcdftime

       
    
    ############ high res :
    data = nc.Dataset(input_file) 
#-----------------------TRAIN
    array = data.variables[var][:] *alpha# assuming “var” is the name of the variable
    print(array.shape)
    if reduce:
        #r1=1
        #r2=2
        #r3=-2
        #r4=-2
        #r5=-4
        r1=1 # skip the last point to kipp the factor 4 of downscaling between the RCM and its upscaled !
        r2=0
        r3=0
        r4=-1
        r5=0
    else:
        r1=0
        r2=1
        r3=array.shape[2]
        r4=0
        r5=-1 
    #print(r1,r2,r3,r4,r5)

    #array_train = np.reshape(array[:,r1:array.shape[1]-r1,r2:r3], (array.shape[0], 1, 1, array.shape[1]+r4, array.shape[2]+r5)) # or use np.expand_dims
    array_train = np.reshape(array[:,:,0:r4], (array.shape[0], 1, 1, array.shape[1], array.shape[2]+r4)) # or use np.expand_dims
    print(array_train.shape)

    tensor = torch.from_numpy(array_train)
    file1 = output_file+".pt"
    torch.save(tensor, file1) # save the tensor to the file
    # saving the time for overlapping 
    times = data.variables["time"]
    times1 = np.asarray(times) # convert to numpy array
    times1 = times1.reshape(-1) # reshape to 1D array
    # create a netcdftime object with the units and calendar of file1
    utime = netcdftime.utime(times.units, calendar=times.calendar)
    times = utime.num2date(times1[:])
    times = [d.date() for d in times] # get only the date components
    torch.save(times, 'time_'+file1) # save the tensor to the file
```

```python
# make target files 
for i in range(1):
    print(i, rcms[i], gcms[i])

    read_netcdf_files(data_dir+rcms[i],rcms[i]+"_target","TOT_PREC",alpha=1, reduce=True)
    read_netcdf_files(data_dir+gcms[i],gcms[i]+"_input","pr",alpha=86400, reduce=True)
    
    print("--------------------------------")
    print(' ')
    print(' ')
```

```python
import os 
import torch
import numpy as np

if  os.path.isfile(rcms[0]+'_target.pt'):
    t1 = torch.load(rcms[0]+'_target.pt')
if  os.path.isfile('time_'+rcms[0]+'_target.pt'):
    time_t1 = torch.load('time_'+rcms[0]+'_target.pt')

if  os.path.isfile(gcms[0]+'_input.pt'):
    tt1 = torch.load(gcms[0]+'_input.pt')
if  os.path.isfile('time_'+gcms[0]+'_input.pt'):
    time_tt1 = torch.load('time_'+gcms[0]+'_input.pt')
    # only the common times !!!
    common_indices = np.intersect1d(time_t1[:], time_tt1[:], return_indices=True) # find the common values and indices of time1 and time2
    #print(common_indices)
    t1_data = t1[common_indices[1], :, :] # slice var1 using the common indices of time1
    tt1_data = tt1[common_indices[2], :, :] # slice var2 using the common indices of time2


import gc 
gc.collect()
del common_indices, time_t1
gc.collect()
del time_tt1
gc.collect()
del t1
gc.collect

del tt1
gc.collect


# if all the gcms uncomment below: 

merged_y = t1_data
merged_X = tt1_data
print(' ')
del t1_data
gc.collect()
del tt1_data
gc.collect()
print(merged_X.shape, merged_y.shape)
```

```python
# use always the clone() command otherwise slicing does not work!!!!

X_train = merged_X[0,:,:,:,:].clone()
y_train = merged_y[0,:,:,:,:].clone()

X_valid = merged_X[1:,:,:,:,:].clone()
y_valid = merged_y[1:,:,:,:,:].clone()

X_test = merged_X[2:,:,:,:,:].clone()
y_test = merged_y[2:,:,:,:,:].clone()


X_train = X_train.float()
X_test = X_test.float()
X_valid = X_valid.float()

y_train = y_train.float()
y_test = y_test.float()
y_valid = y_valid.float()

print(' ')

# save final files to be fed into the CNN model: 
torch.save(y_train, 'target_train.pt')
torch.save(y_valid, 'target_val.pt')
torch.save(y_test, 'target_test.pt')



torch.save(X_train, 'input_train.pt')
torch.save(X_valid, 'input_val.pt')
torch.save(X_test, 'input_test.pt')
```
